```python
import pandas as pd
import networkx as nx
```

```python
# Load the dataset
G = pd.read_csv(r"C:\Users\Swapnil Mishra\Desktop\DS_Codes_Swapnil\Network_Analytics\routes.csv")
G = G.iloc[0:500,1:10]
G.info()
```

```python
g = nx.Graph()
g = nx.from_pandas_edgelist(G,source = 'Source Airport',target = 'Destination Airport')
```

```python
print(g)
```

```python
# Degree Centrality
d = nx.degree_centrality(g)
print(d)
```

```python
pos = nx.spring_layout(g)
nx.draw_networkx(g, pos, node_size = 15, node_color = 'red')
```

```python
# closeness centrality
closeness = nx.closeness_centrality(g)
print(closeness)
```

```python
# Betweeness Centrality
b = nx.betweenness_centrality(g)
print(b)
```

```python
# Eigen-Vector Centrality
evg = nx.eigenvector_centrality(g)
print(evg)
```

```python
# Cluster coefficient
cluster_coeff = nx.clustering(g)
print(cluster_coeff)
```

```python
# Average clustering
cc = nx.average_clustering(g)
print(cc)
```
